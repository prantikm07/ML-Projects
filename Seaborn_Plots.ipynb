```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Categorical plot** -- at least one cat col

- count plot
- box plot
- violin plot
- bar plt
- swarm plot
- strip plot

**Distribution Plot** -- at least one num col

- dis/dist plot
- hist plot
- kde plot
- rug plot
- join plot

**regression plot**    - at least 2 num col

- scatter plot
- reg plot
- 1m plot
- rel plot
- cat plot

**Matrix Plot** --- cat(num) col

- pair plot
- heatmap

```python
df = sns.load_dataset('tips')
df.head(10)
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(x='sex', data = df)
```

```python
sns.countplot(x='sex', data = df, hue='smoker', palette='Set2')
```

```python
sns.boxplot(x='day', y = 'total_bill', data=df, palette='husl', hue='sex')
```

```python
sns.violinplot(x='day', y = 'total_bill', data=df, palette='husl')
```

```python
sns.swarmplot(x='day', y = 'total_bill', data=df, palette='husl', hue='sex')
```

```python
import warnings
warnings.filterwarnings('ignore') # -- hide all warnings
```

```python
sns.stripplot(x='day', y = 'total_bill', data=df, palette='husl', hue='sex')
```

```python
sns.barplot(x='day', y = 'total_bill', data=df, palette='husl', hue='sex')
```

```python
sns.barplot(x='day', y = 'total_bill', data=df, palette='husl', estimator='sum')
```

```python
sns.displot(x='total_bill', data=df)
```

```python
sns.displot(x='total_bill', data=df, kind='kde')
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.displot(x='tip', data=df, palette='Set2', kind='kde', hue='sex')
```

```python
sns.kdeplot(x='tip', data=df, palette='Set2', hue='sex')
```

```python
sns.kdeplot(x='total_bill', data=df)
sns.rugplot(x='total_bill', data=df)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde', color='red', fill='true')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex', color='red')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex', style='smoker', size='time')
```

```python
sns.regplot(x='total_bill',y='tip', data=df)
```

```python
sns.lmplot(x='total_bill',y='tip', data=df)
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', style='smoker', size='time')
```

```python
sns.catplot(x='time',y='total_bill', data=df, kind='swarm', hue='sex', col='day', aspect=0.5)
```

```python
sns.pairplot(df, hue='smoker')
```
