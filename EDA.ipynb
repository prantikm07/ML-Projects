```python
x =np.linspace(0, 10, 1000)
plt.figure(figsize=(10, 4))
plt.plot(x, np.sin(x), label="Sin Curve")
plt.plot(x, np.cos(x), label="Cos Curve")
plt.legend(loc="upper right")
plt.show()
```

```python
# sns.get_dataset_names()
df1 = sns.load_dataset('tips')
df1
```

```python
plt.figure(figsize=(15,4))
plt.plot(df1['total_bill'], color = 'r', lw = 3, ls = '-', marker='*', ms=12, mfc='y', mec='g')
plt.title('Line plot of price')
plt.xlabel("X -- axis")
plt.ylabel("Price")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1, 2, 1)
plt.plot(df1['total_bill'], color = 'r', lw = 3, ls = '-', marker='*', ms=8, mfc='y', mec='g')
plt.title('Line plot of Total Bill')
plt.xlabel("X -- axis")
plt.ylabel("Price")
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(df1['tip'], color = 'g', lw = 3, ls = '-', marker='*', ms=8, mfc='y', mec='r')
plt.title('Line plot of Tips')
plt.xlabel("X -- axis")
plt.ylabel("Price")
plt.grid()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(x = 'day', height= 'total_bill', data = df1, color = 'magenta')
plt.xlabel("Days")
plt.ylabel("Bill")
```
