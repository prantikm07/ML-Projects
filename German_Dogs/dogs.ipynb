```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("data_02.csv")
data
```

```python
data.describe()
```

```python
class_id = {'Dobermann': 0, 'German Shepherd': 1, 'Rottweiler': 2}
data['Class ID'] = data['Class'].apply(lambda x: class_id[x])
data
```

```python
fig, ax = plt.subplots()
ax.scatter(x=data['Weight'], y=data['Height'],c=data['Class ID'])
```

```python
model = svm.SVC(kernel='linear')
x = data[['Weight', 'Height']]
y = data['Class ID']
model.fit(x,y)
```

```python
x_test = np.random.rand(10000, 2)
x_test = x_test*(45,20) + (25, 55)
y_pred = model.predict(x_test)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=x_test[:,0],y=x_test[:,1], c=y_pred, alpha=.5)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x=x_test[:,0],y=x_test[:,1], c=y_pred, alpha=.5)
colors= ['r', 'g', 'b']
ax.scatter(x=data['Weight'], y=data['Height'],c=data['Class ID'].apply(lambda x: colors[x]))
plt.show()
```
